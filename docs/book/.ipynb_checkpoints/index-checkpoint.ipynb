# OpenFisca-UK

This book contains an introduction to using OpenFisca-UK to model UK taxes and benefits.

## Short demo

Calculating, for example, the average National Insurance liability by region can be done with the following code:

```python
from openfisca_uk import PopulationSim
from openfisca_uk.variables.household.attributes import Region
import pandas as pd
import numpy as np
from rdbl import gbp

sim = PopulationSim()
df = pd.DataFrame()
df["Region"] = sim.calc("region", period="2020").decode_to_str()

df["Population"] = (
    sim.calc("people_in_household") 
    * sim.calc("household_weight")
)
df["NI"] = (
    sim.calc("NI", map_to="household", period="2020") 
    * sim.calc("household_weight")
)

df = df.groupby("Region").sum()

df["NI per capita"] = df["NI"] / df["Population"]
df["Population"] = df["Population"].apply(gbp)
df["NI"] = df["NI"].apply(gbp)
df["NI per capita"] = df["NI per capita"].apply(gbp)
df
```
